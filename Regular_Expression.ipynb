```python
import re
```

```python
text = ''' text (999)-999-9999 text 1009998888 '''
pattern = '\(\d{3}\)-\d{3}-\d{4} | \d{10}'

matches = re.findall(pattern, text)
matches
```

```python
text2 = ''' 
note 1 : Overview
description
note 2 : context
context
note 3 : product detail
detail
note 4 : end
close
'''
# pattern2 = 'note \d : [^\n]+'
pattern2 = 'note \d : ([^\n]*)'

matches2 = re.findall(pattern2, text2)
matches2
```

```python
text3 = ''' 
FY2026 Q1
Profit $4.2 billions
FY2026 Q2
Profit $3 billions
FY2026 Q3
Profit $2.2 billions
FY2026 Q4
Profit $2 billions
'''
pattern3 = 'FY\d{4} Q[1-4]'

matches3 = re.findall(pattern3, text3,flags=re.IGNORECASE)
matches3
```

```python
pattern4 = 'FY(\d{4} Q[1-4])'

matches4 = re.findall(pattern4, text3,flags=re.IGNORECASE)
matches4
```

```python
pattern5 = '\$([0-9\.]+)'

matches5 = re.findall(pattern5, text3,flags=re.IGNORECASE)
matches5
```

```python
pattern6 = 'FY(\d{4} Q[1-4])[^\$]+\$([0-9\.]+)'

matches6 = re.findall(pattern6, text3,flags=re.IGNORECASE)
matches6
```

```python
search1 = re.search(pattern6, text3)
search1.groups()
```
