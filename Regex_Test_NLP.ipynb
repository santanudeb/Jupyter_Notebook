```python
import re
```

```python
chat1 = 'codebasics: you ask lot of questions. 😠. 1235678912, abc@xyz.com'
chat2 = 'codebasics: here it is: (123)-567-8912, abc@xyz.com'
chat3 = 'codebasics: yes, phone: 1235678912 email: abc@xyz.com'
```

```python
pattern_phone_number = '\d{10}|\(\d{3}\)-\d{3}-\d{4}'

phone_find = re.findall(pattern_phone_number, chat2)
phone_find
```

```python
pattern_email = '[a-zA-Z0-9_]*@[a-zA-Z]*\.[a-zA-Z]*'
email_find = re.findall(pattern_email, chat2)
email_find
```
