```python
# Merging DataFrames in Pandas
```

```python
df1_read_csv_lotr = pd.read_csv(r'C:\Users\santa\Downloads\Panda Test\LOTR.csv')
df1_read_csv_lotr
```

```python
df2_read_csv_lotr = pd.read_csv(r'C:\Users\santa\Downloads\Panda Test\LOTR 2.csv')
df2_read_csv_lotr
```

```python
df1_read_csv_lotr.merge(df2_read_csv_lotr)
```

```python
df1_read_csv_lotr.merge(df2_read_csv_lotr, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1_read_csv_lotr.merge(df2_read_csv_lotr, how = 'outer')
```

```python
df1_read_csv_lotr.merge(df2_read_csv_lotr, how = 'left')
```

```python
df1_read_csv_lotr.merge(df2_read_csv_lotr, how = 'right')
```

```python
df1_read_csv_lotr.merge(df2_read_csv_lotr, how = 'cross')
```

```python
df1_read_csv_lotr.join(df2_read_csv_lotr, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
df1_joins_df2 = df1_read_csv_lotr.set_index('FellowshipID').join(df2_read_csv_lotr.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right')
df1_joins_df2
```

```python
df1_joins_df2 = df1_read_csv_lotr.set_index('FellowshipID').join(df2_read_csv_lotr.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df1_joins_df2
```

```python
pd.concat([df1_read_csv_lotr, df2_read_csv_lotr])
```

```python
pd.concat([df1_read_csv_lotr, df2_read_csv_lotr], join = 'inner')
```

```python
pd.concat([df1_read_csv_lotr, df2_read_csv_lotr], join = 'outer')
```

```python
pd.concat([df1_read_csv_lotr, df2_read_csv_lotr], join = 'outer', axis = 1)
```
