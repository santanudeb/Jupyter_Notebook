```python
weight = int(input('Enter weight : '))
height = int(input('Enter height : '))
bmi = (weight * 703) / (height * height)
print('your bmi is : ',bmi)
```

```python
if bmi < 18.5 :
    print('Weight type : under weight & health risks : minimal')
elif bmi >= 18.5 and bmi <= 24.9 :
    print('Weight type : normal weight & health risks : minimal')
elif bmi >= 25 and bmi <= 29.9 :
    print('Weight type : over weight & health risks : increased')
elif bmi >= 30 and bmi <= 34.9 :
    print('Weight type : obese & health risks : high')
elif bmi >= 35 and bmi <= 39.9 :
    print('Weight type : severely obese & health risks : very high')
elif bmi >= 40 :
    print('Weight type : morbidly obese & health risks : extremely high')
else :
    print('not found')
```

```python
import os, shutil
path = r"C:/Users/santa/Downloads/Python Test/"
file_names = os.listdir(path)
```

```python
FolderNames = ['txt files', 'pdf files', 'image files']

for i in range (0,3) :
    if not os.path.exists(path + FolderNames[i]):
        print(path + FolderNames[i])
        os.makedirs(path + FolderNames[i])
```

```python
for file in file_names :
    if ".txt" in file and not os.path.exists(path + "txt files/" + file):
        shutil.move(path + file, path + "txt files/" + file)
    elif ".pdf" in file and not os.path.exists(path + "pdf files/" + file):
        shutil.move(path + file, path + "pdf files/" + file)
    elif ".jpg" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
```
